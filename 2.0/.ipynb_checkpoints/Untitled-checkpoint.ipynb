```python
from datetime import datetime 
from datetime import timedelta
import sqlite3 as lite
import time
import jieba
import time
con = lite.connect('/ods/interns/interns/J0186/project_flask/FBWEB.sqlite')
c = con.cursor()
headers = {'User-Agent' :'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.99 Safari/537.36'}
```

```python
today = datetime.today()
today = datetime.strftime(today, '%Y/%m/%d')
#today = '2018-07-31'
```

```python
summary = c.execute("SELECT content FROM News WHERE Id > 1 AND Id < 100").fetchall()
print(summary)
```

```python
start_time = time.time()
stopWords = []
words_ary = []
re = {}
summation = 0
counter = 0
strsummary = str(summary)
with open('stopWords.txt', 'r', encoding='UTF-8') as file:
    for data in file.readlines():
        data = data.strip()
        stopWords.append(data)

for word in jieba.cut(strsummary):
    words_ary.append(word)
    from collections import Counter
    words_ary = [word for word in jieba.cut(strsummary)]
    c = Counter(words_ary)
    hik ={}
    for k, v in c.most_common(100):
        if k in stopWords:
            continue
        else:
            if len(k) >= 2:##k = 我、的(少於一個字的詞) V = 詞的最低數量
                #cur.execute("INSERT INTO wordsCount (words,wordsCount,daytime,frequencyNum)VALUES (?,?,?,?)", (k, v, today, 1))
                hik[k]=hik.get(k,v)
                summation = summation + hik[k]
                counter = counter + 1

flag = summation / counter
for h in hik:
    if(hik[h] > flag or hik[h] == flag):
        re[h] = hik[h]
print(re)
print("--- %s seconds ---" % (time.time() - start_time))
```
